```python
for county in countyData: 
#  STEP 1: Grab data
    # Var 'StateCountyID' using data[i].fips (SELECT StateCountyID FROM tblStateCounty WHERE FIPS == data[i].fips)
    fips = county["fips"]
    print("Fips:", fips)
    
    # Var 'Uploaded' DATETIME (Should be current date)
    dateValue = datetime.datetime.now()
    print("Uploaded:", dateValue)
    #  Var 'PosTestRateCounty'
    PosTestRateCounty = county["metrics"]["testPositivityRatio"]
    if PosTestRateCounty == None:
        stateAcr = county["state"]
        for state in stateData:
            if state["state"] == stateAcr:
                PosTestRateCounty = state["metrics"]["testPositivityRatio"]
    print("PosTestRateCounty:", PosTestRateCounty)
        # var postestratecounty;
    
    # Var 'NumNewCases' 
    numNewCases = county["actuals"]["newCases"]
    print("numNewCases:",numNewCases)
    
# STEP 2: Insert data into db
```
